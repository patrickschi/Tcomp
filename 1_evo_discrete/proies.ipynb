```python
import matplotlib.pyplot as plt

x ,y  = 1450 ,55      # Nombre initial de proies et de prédateurs
n = 5000             # nombre d'années

vx, vy = [x], [30*y]
for i in range(n) :
    x ,y = 1.05*x - 0.001*x*y ,0.00002*x*y + 0.97*y
    vx.append(x)
    vy.append(30*y)

vn=[i for i in range(n+1)]
    
# graphique
plt.scatter(vn, vx, color='g', label='proies')
plt.scatter(vn, vy, color='r', label='30 * prédateurs')
plt.legend()
plt.show()
```
